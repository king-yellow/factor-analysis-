```python
import numpy as np
import random

def estimate_factor(factor,F0,F0_grow):    #only can apply to SMB, RMW, CMA 
    
    if factor=="SMB":
        Range=0.477027
        threshold0=75
        threshold1=81
    if factor=="RMW":
        Range=0.583761
        threshold0=90
        threshold1=83
    if factor=="CMA":
        Range=0.322870
        threshold0=81
        threshold1=67
        
    if F0_grow>0 or F0_grow==0:
        threshold=threshold0
    else:
        threshold=threshold1
    y=range(1,100,1)
    #random.seed(1)
    number=random.sample(y,1)
    
    threshold=[threshold]
    
    if number<threshold or number==threshold:
        if F0_grow>0:
            direction=-1
        else:
            direction=1
    else:
        if F0_grow>0:
            direction=1
        else:
            direction=-1
    F1=F0+F0*Range*direction     
    return (F1)

def direction(factor,F0,F0_grow):    #only can apply to SMB, RMW, CMA 
    
    if factor=="SMB":
        Range=0.477027
        threshold0=75
        threshold1=81
    if factor=="RMW":
        Range=0.583761
        threshold0=90
        threshold1=83
    if factor=="CMA":
        Range=0.322870
        threshold0=81
        threshold1=67
        
    if F0_grow>0 or F0_grow==0:
        threshold=threshold0
    else:
        threshold=threshold1
    y=range(1,100,1)
    #random.seed(1)
    number=random.sample(y,1)
    
    threshold=[threshold]
    
    if number<threshold or number==threshold:
        if F0_grow>0:
            direction=-1
        else:
            direction=1
    else:
        if F0_grow>0:
            direction=1
        else:
            direction=-1
    F1=F0+F0*Range*direction     
    return (direction)

def accuracy(gama):      
    u=0
    l=0
    while u<1000:

        SMB0=x[1:47,0]   
        SMB1=x[2:47,0]  #actual SMB1
        DSMB=x[1:47,7]
        E=np.array([0])
        F=np.array([0])

        for i in range(0,46):
            F0=float(SMB0[i])
            F0_grow=float(DSMB[i])
            Estimated_SMB=estimate_factor("SMB",F0,F0_grow)
            Estimated_direction=direction("SMB",F0,F0_grow)


            E=np.row_stack((E, [Estimated_SMB]))
            F=np.row_stack((F, [Estimated_direction]))
        #print("Estimated_SMB=",E[1:46,:])



        together=np.column_stack((E[1:46,:], SMB1,F[1:46,:],x[1:46,0]))
        #print(together)

        h=0
        hg=0
        for j in range(0,45):
            T=together[j][0]*together[j][1]
            TT=together[j][2]*together[j][3]
            if TT>0 or abs(together[j][0])>gama:
                hg=hg+1
                if T>0:
                    h=h+1

        accuracy=h/hg
        #print(accuracy)
        l=l+accuracy
        u=u+1

    mean_acciracy=l/1000

    return(mean_acciracy)

def trade_time(gama):      
    u=0
    l=0
    while u<1000:

        SMB0=x[1:47,0]   
        SMB1=x[2:47,0]  #actual SMB1
        DSMB=x[1:47,7]
        E=np.array([0])
        F=np.array([0])

        for i in range(0,46):
            F0=float(SMB0[i])
            F0_grow=float(DSMB[i])
            Estimated_SMB=estimate_factor("SMB",F0,F0_grow)
            Estimated_direction=direction("SMB",F0,F0_grow)


            E=np.row_stack((E, [Estimated_SMB]))
            F=np.row_stack((F, [Estimated_direction]))
        #print("Estimated_SMB=",E[1:46,:])



        together=np.column_stack((E[1:46,:], SMB1,F[1:46,:],x[1:46,0]))
        #print(together)

        h=0
        hg=0
        for j in range(0,45):
            T=together[j][0]*together[j][1]
            TT=together[j][2]*together[j][3]
            if TT>0 or abs(together[j][0])>gama:
                hg=hg+1
                if T>0:
                    h=h+1

        #accuracy=h/hg
        #print(accuracy)
        l=l+hg
        u=u+1

    mean_hg=l/1000

    return(mean_hg)

z=np.array([0])
v=np.array([0])
zz=np.array([0])
vv=np.array([0])
gama=0.01
while gama<0.05:
    q=accuracy(gama)
    t=trade_time(gama)
    f=t*q
    zz=np.row_stack((zz, q))
    vv=np.row_stack((vv, t))
    z=np.row_stack((z, gama))
    v=np.row_stack((v, f))
   
    gama=gama+0.001

s=np.column_stack((z, zz,vv,v))
print(s)
    
```
