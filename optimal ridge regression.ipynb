```python
x1=x[:,0:7]
#print("x=",x1)
y=x[:,7]
#print("y=",y)

%matplotlib inline
lambdda=0.008

import numpy as np
def least_squares_weight(x, y):
    
    xTx = x.T.dot(x)
    #print(xTx.shape,y.shape)
    I=lambdda*np.eye(7, dtype=int)
    xTx_inv = np.linalg.inv(xTx+I)
    w = xTx_inv.dot(x.T.dot(y))
    y_hat = x.dot(w)
    #print("y=",y,"y_hat=",y_hat)
    error = np.mean((y - y_hat) ** 2)
    return w

def least_squares_error(x, y):
    
    xTx = x.T.dot(x)
    #print(xTx.shape,y.shape)
    I=lambdda*np.eye(7, dtype=int)
    xTx_inv = np.linalg.inv(xTx+I)
    w = xTx_inv.dot(x.T.dot(y))
    y_hat = x.dot(w)
    #print("y=",y,"y_hat=",y_hat)
    error = np.mean((y - y_hat) ** 2)
    return error

def least_squares_yhat(x, y):
    
    xTx = x.T.dot(x)
    #print(xTx.shape,y.shape)
    I=lambdda*np.eye(7, dtype=int)
    xTx_inv = np.linalg.inv(xTx+I)
    w = xTx_inv.dot(x.T.dot(y))
    y_hat = x.dot(w)
    #print("y=",y,"y_hat=",y_hat)
    error = np.mean((y - y_hat) ** 2)
    return y_hat

#print(least_squares(x1, y))

a=0
em=np.array([0])
Rm=np.array([0])
wm=np.array([0,0,0,0,0,0,0])

for i in range(7,1104):
    y=x[:,i]
    y_var=np.var(y)
    w=least_squares_weight(x1, y)
    wm=np.vstack((wm,w))
    e=least_squares_error(x1, y)
    residule=y-least_squares_yhat(x1, y)
    #print(residule)
    R_square=1-(np.sum(residule**2)/(47*y_var))
    Rm=np.vstack((Rm,R_square))
    em=np.vstack((em,e))
    a=a+1
print(a)
error=em[1:1098,:]
print("error=",error)
weight=wm[1:1098,:]
print("weight=",weight.T)
R_square=Rm[1:1098,:]
print("R_square=",R_square)

beta=0
import matplotlib.pyplot as plt
W1=weight[0]
W1_mean=np.mean(W1)
if W1_mean>0:
    beta=beta+1
print("W1_mean=",W1_mean)
plt.title('SMB')
plt.xlabel("beta")
plt.ylabel("frequency")
plt.hist(W1,bins=10,normed=False)

plt.show()

W2=weight[1]
W2_mean=np.mean(W2)
if W2_mean>0:
    beta=beta+1
print("W2_mean=",W2_mean)
plt.title('RMW')
plt.xlabel("beta")
plt.ylabel("frequency")
plt.hist(W2,bins=10,normed=False)
plt.show()

W3=weight[2]
W3_mean=np.mean(W3)
if W3_mean>0:
    beta=beta+1
print("W3_mean=",W3_mean)
plt.title('HML')
plt.xlabel("beta")
plt.ylabel("frequency")
plt.hist(W3,bins=10,normed=False)
plt.show()

W4=weight[3]
W4_mean=np.mean(W4)
if W4_mean>0:
    beta=beta+1
print("W4_mean=",W4_mean)
plt.title('CMA')
plt.xlabel("beta")
plt.ylabel("frequency")
plt.hist(W4,bins=10,normed=False)
plt.show()

W5=weight[4]
W5_mean=np.mean(W5)
if W5_mean>0:
    beta=beta+1
print("W5_mean=",W5_mean)
plt.title('Rm-Rf')
plt.xlabel("beta")
plt.ylabel("frequency")
plt.hist(W5,bins=10,normed=False)
plt.show()

W6=weight[5]
W6_mean=np.mean(W6)
if W6_mean>0:
    beta=beta+1
print("W6_mean=",W6_mean)
plt.title('CRMHL')
plt.xlabel("beta")
plt.ylabel("frequency")
plt.hist(W6,bins=10,normed=False)
plt.show()

W7=weight[6]
W7_mean=np.mean(W7)
if W7_mean>0:
    beta=beta+1
print("W7_mean=",W7_mean)
plt.title('AMLH')
plt.xlabel("beta")
plt.ylabel("frequency")
plt.hist(W7,bins=10,normed=False)
plt.show()

R_square_mean=np.mean(R_square)
print("R_square_mean=",R_square_mean)
plt.title('R_square')
plt.xlabel("beta")
plt.ylabel("frequency")
plt.hist(R_square,bins=10,normed=False)
plt.show()


#################################################################
vx=np.array([0])
tx=np.array([0])
for i in range(0,7):
    vn=x[:,i]
    #print(vn)
    vn_var=np.var(vn)*47
    vx=np.vstack((vx,vn_var))
    #print("vn_var=",vn_var)
#print("vx=",vx[1:8])  

x1=x[:,0:7]  
for i in range(7,1104):
    y=x[:,i]
    
    weight=wm[1:1098,:]
    
    #print("weight=",weight[i-7,:])
    #print("y=",y)
    residule=y-least_squares_yhat(x1, y)
    y_hat2=(np.sum(residule)**2)**2
    y_hat3=y_hat2/39

    y_bar2=np.var(y)*47
    divide=np.sqrt(y_hat3/(vx[1]*y_bar2))

    for j in range(0,7):
        #print("weight=",weight[i-7,j])
        t=weight[0,j]/divide
        #print("t=",t)
        tx=np.vstack((tx,t))
            
#print("tx=",tx[1:7680])
#print("tx=",tx.shape)

##############################################################
new_t=tx[1:7680]
t_value=new_t.reshape((1097,7))
#print(t_value)
#print(t_value.shape)

new_t=tx[1:7680]
p_value=[0]
for i in range(0,7679):
    if new_t[i]>2.021 or new_t[i]<-2.021:
        p=0.05
    else:
        p=1
    p_value=np.vstack((p_value,p))
#print(p_value)


real_p_value=p_value[1:7680]
#print(real_p_value)  

p_value=real_p_value.reshape((1097,7))
#print(p_value[:,0])
#print(p_value.shape)
P=0
P2=0
for u in range(0,7):
    a=0
    for k in p_value[:,u]:
        if k==0.05:
            a=a+1
        else:
            a=a
    p_level=a/1097
    P=P+p_level
    if p_level>0.9:
        P2=P2+1
    print("p_level=",p_level)
P_average_value=P/7

print("lamda=",lambdda)
print("P_average_value",P_average_value)
print("beta=",beta)
Q=P_average_value+beta+P2
print("Q=",Q)
```
