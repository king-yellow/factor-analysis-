```python
import numpy as np
import random


def estimate_index(STD,index0,index0_grow,threshold0,threshold1):    #only can apply to SMB, RMW, CMA 
    a=np.array([0])
    b=np.array([0])
    if index0_grow>0 or index0_grow==0:
        threshold=threshold0
    else:
        threshold=threshold1

    
    CI=np.random.normal(loc=0, scale=STD, size=1000)
    #CI=[CI]
    #print(CI)
    for i in CI:
        #print(i)
        if i>0 or i==0:
            a=np.row_stack((a, [i]))
            #print("a=",a)
        else:
            b=np.row_stack((b, [i]))
    print("a=",a)
    print("b=",b)
    x=range(1,100,1)
    number=random.sample(x,1)
    
    #number=str(number)
    threshold=[threshold]
    
    #print("random number=",number)
    #print("threshold=",threshold)
    #number2=str(number)
    #print(type(index0_grow))
    
    
    
    if number<threshold or number==threshold:
        if index0_grow>0:
            rate=b[number,0]
            #print("b")
        else:
            rate=a[number,0]
            #print("a")
        #rate=str(rate)
        #print(rate)
    if number>threshold:
        if index0_grow>0:
            rate=a[number,0]
            #print("a2")
        else:
            rate=b[number,0]
            #print("b2")
        #rate=str(rate)
        #print(rate)
    rate=float(rate)
    #print("flutuation rate=",rate)
    estimate_index=index0+rate
    #print("estimate index=",estimate_index)

    return(estimate_index)

#estimate_index(0.09824863187300051,-0.4409948776066145,-0.6087774755459053,75,81)


def estimate_index_rate(STD,index0,index0_grow,threshold0,threshold1):    #only can apply to SMB, RMW, CMA 
    a=np.array([0])
    b=np.array([0])
    if index0_grow>0 or index0_grow==0:
        threshold=threshold0
    else:
        threshold=threshold1

    
    CI=np.random.normal(loc=0, scale=STD, size=1000)
    #CI=[CI]
    #print(CI)
    
    for i in CI:
        #print(i)
        if i>0 or i==0:
            a=np.row_stack((a, [i]))
            #print("a=",a)
        else:
            b=np.row_stack((b, [i]))
    #print("a=",a)
    #print("b=",b)
    x=range(1,100,1)
    number=random.sample(x,1)
    
    #number=str(number)
    threshold=[threshold]
    #print("random number=",number)
    #print("threshold=",threshold)
    #number2=str(number)
    #print(type(index0_grow))
    
    
    
    if number<threshold or number==threshold:
        if index0_grow>0:
            rate=b[number,0]
            #print("b")
        else:
            rate=a[number,0]
            #print("a")
        #rate=str(rate)
        #print(rate)
    if number>threshold:
        if index0_grow>0:
            rate=a[number,0]
            #print("a2")
        else:
            rate=b[number,0]
            #print("b2")
        #rate=str(rate)
        #print(rate)
    rate=float(rate)
    #print("flutuation rate=",rate)
    estimate_index=index0+rate
    #print("estimate index=",estimate_index)

    return(rate)

estimate_index(0.09824863187300051,-0.4409948776066145,-0.6087774755459053,75,81)

```

```python
def estimate_index_HML(STD,index0,index0_grow,index1_grow,threshold00,threshold11,threshold10,threshold01):    
    if index0_grow>0 or index0_grow==0:
        if index1_grow>0 or index1_grow==0:
            threshold=threshold00
        else:
            threshold=threshold10
    else:
        if index1_grow>0 or index1_grow==0:
            threshold=threshold01
        else:
            threshold=threshold11
        
        
    a=np.array([0])
    b=np.array([0])
    CI=np.random.normal(loc=0, scale=STD, size=1000)
    #CI=[CI]
    #print(CI)
    for i in CI:
        #print(i)
        if i>0 or i==0:
            a=np.row_stack((a, [i]))
            
        else:
            b=np.row_stack((b, [i]))
    #print("a=",a)
    #print("b=",b)
    x=range(1,100,1)
    number=random.sample(x,1)
    
    #number=str(number)
    threshold=[threshold]
    print("random number=",number)
    print("threshold=",threshold)
    #number2=str(number)
    #print(type(index0_grow))
    
    #only for AMLH-forecast HML
    if threshold==threshold10:
        if number<threshold or number==threshold:
            if index0_grow>0:
                rate=a[number,0]
            else:
                rate=b[number,0]

        if number>threshold:
            if index0_grow>0:
                rate=b[number,0]
            else:
                rate=a[number,0]
        
    else:
        if number<threshold or number==threshold:
            if index0_grow>0:
                rate=b[number,0]
            else:
                rate=a[number,0]

        if number>threshold:
            if index0_grow>0:
                rate=a[number,0]
            else:
                rate=b[number,0]

    rate=float(rate)
    print("flutuation rate=",rate)
    estimate_index=index0+rate
    print("estimate index=",estimate_index)
    return (estimate_index)

```

```python
def estimate_index_RMRF(STD,index0,index0_grow,index1_grow,index2_grow,threshold000,threshold011,threshold010,threshold001,threshold100,threshold111,threshold110,threshold101):    
    if index0_grow>0 or index0_grow==0:
        if index1_grow>0 or index1_grow==0:
            if index2_grow>0 or index1_grow==0:
                threshold=threshold000
            else:
                threshold=threshold010
        else:
            if index2_grow>0 or index1_grow==0:
                threshold=threshold100
            else:
                threshold=threshold110
            
    else:
        if index1_grow>0 or index1_grow==0:
            if index2_grow>0 or index1_grow==0:
                threshold=threshold001
            else:
                threshold=threshold011
        else:
            if index2_grow>0 or index1_grow==0:
                threshold=threshold101
            else:
                threshold=threshold111
        
        
    a=np.array([0])
    b=np.array([0])
    
    CI=np.random.normal(loc=0, scale=STD, size=1000)
    #CI=[CI]
    #print(CI)
    for i in CI:
        #print(i)
        if i>0 or i==0:
            a=np.row_stack((a, [i]))
            
        else:
            b=np.row_stack((b, [i]))
    #print("a=",a)
    #print("b=",b)
    x=range(1,100,1)
    number=random.sample(x,1)
    
    #number=str(number)
    threshold=[threshold]
    print("random number=",number)
    print("threshold=",threshold)
    #number2=str(number)
    #print(type(index0_grow))
    
    #only for Rm-Rf forecasting
    
    if number<threshold or number==threshold:
        if index0_grow>0:
            rate=a[number,0]
        else:
            rate=b[number,0]

    if number>threshold:
        if index0_grow>0:
            rate=b[number,0]
        else:
            rate=a[number,0]
        

    rate=float(rate)
    print("flutuation rate=",rate)
    estimate_index=index0+rate
    print("estimate index=",estimate_index)
    return (estimate_index)
```

```python
SMB=x[1:47,0]
print(SMB)
DSMB=x[1:47,7]
STD=float(np.sqrt(np.var(SMB)))
#print(STD)
threshold0=75
threshold1=81
E=np.array([0])
R=np.array([0])
for i in range(0,46):
    index0=float(SMB[i])
    #print(index0)
    index0_grow=float(DSMB[i])
    #print(index0_grow)
    Estimated_SMB=estimate_index(STD,index0,index0_grow,threshold0,threshold1)
    Estimated_SMB_rate=estimate_index_rate(STD,index0,index0_grow,threshold0,threshold1)
    E=np.row_stack((E, [Estimated_SMB]))
    R=np.row_stack((R, [Estimated_SMB_rate]))
    
E=E[1:46,:]
R=R[1:46,:]
print("Estimated_SMB=",E)
print("Estimated_SMB_rate=",R)
SMB2=x[1:46,0]           #real SMB
SMB2_rate=x[2:47,7]      #real SMB flutuate rate
print("SMB2=",SMB2)
print("SMB2_rate=",SMB2_rate)

SMB_total= np.column_stack((E,R,SMB2,SMB2_rate))


print("estimated SMB, forecast flutuated rate,real SMB,real SMB flutuate rate=",SMB_total)


SMB3=SMB_total[:,1]*SMB_total[:,3]
print(SMB3)
h=0
j=0
for i in SMB3:
    if i>0:
        h=h+1
    else:
        j=j+1
correct_rate=h/(h+j)
print("correct_rate=",correct_rate)   
        
   
```
